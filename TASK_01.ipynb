```python
import matplotlib.pyplot as plt
```

```python
# Bar Chart for Income Distributions by Regions
income_by_region = pd.crosstab(df_meta.IncomeGroup, df_meta.Region)
plt.figure(figsize=(10, 6))
income_by_region.plot(kind='bar')
plt.xlabel("Income Group")
plt.ylabel("Number of Countries in the Regions")
plt.title("Income Distribution by Regions")
plt.legend(title='Regions')
plt.xticks(rotation=45)
plt.show()
```

```python
df.columns
```

```python
#top 10 countries based on population in the year 1960
df_sorted1 = df.sort_values(by='1960', ascending = False).head(10)
plt.figure(figsize=(10,6))
plt.barh(df_sorted1['Country Name'], df_sorted1['1960'], color = 'purple')
plt.xlabel('Population')
plt.title('Top 10 Countries by Population in 1960')
plt.gca().invert_yaxis() 
plt.show()
```

```python
#top 10 countries based on population in the year 2022
df_sorted1 = df.sort_values(by='2022', ascending = False).head(10)
plt.figure(figsize=(10,6))
plt.barh(df_sorted1['Country Name'], df_sorted1['2022'], color = 'orange')
plt.xlabel('Population')
plt.title('Top 10 Countries by Population in 2022')
plt.gca().invert_yaxis() 
plt.show()
```

```python
#bottom 10 countied based on population in the year 1960
df_sorted3 = df.sort_values(by='1960', ascending=False).tail(10)
plt.figure(figsize=(10, 6))
plt.barh(df_sorted3['Country Name'], df_sorted3['1960'], color='green')
plt.xlabel('Population')
plt.title('Bottom 10 Countries by Population in 1960')
plt.gca().invert_yaxis()
plt.show()
```

```python
#bottom 10 countied based on population in the year 2022
df_sorted3 = df.sort_values(by='2022', ascending=False).tail(10)
plt.figure(figsize=(10, 6))
plt.barh(df_sorted3['Country Name'], df_sorted3['2022'], color='darkblue')
plt.xlabel('Population')
plt.title('Bottom 10 Countries by Population in 2022')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Histogram for population distribution in India from 1960-2022
data = df[df['Country Code'] == 'IND']
years = [str(year) for year in range(1960, 2023)]  
population = data[years].values.flatten()
plt.figure(figsize=(10, 6))
plt.hist(population, bins=15, edgecolor = 'black', color = 'skyblue')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Distribution in India (1960-2022)')
plt.show()
```
